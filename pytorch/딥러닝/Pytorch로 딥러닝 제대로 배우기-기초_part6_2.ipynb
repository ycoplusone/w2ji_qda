```python
import os
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
```

### 1. 모델 만들기

#### (1) 모델 구조 정의

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 10),
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
```

#### (2) 모델 인스턴스화

```python
model = NeuralNetwork().to(device)
print(model)
```

### 2. 다양한 모델 구성 방법

#### (1) 하나씩 선언하기

```python
class OneByOneNetowrk(nn.Module):
    def __init__(self):
        super(OneByOneNetowrk, self).__init__()
        self.flatten = nn.Flatten()
        self.linear1 = nn.Linear(28*28, 512)
        self.relu1 = nn.ReLU()
        self.linear2 = nn.Linear(512, 512)
        self.relu2 = nn.ReLU()
        self.linear3 = nn.Linear(512, 10)

    def forward(self, x):
        x = self.flatten(x)
        x = self.linear1(x)
        x = self.relu(x)
        x = self.linear2(x)
        x = self.relu2(x)
        logits = self.linear3(x)
        return logits
```

```python
one_by_one_model = OneByOneNetowrk().to(device)
print(one_by_one_model)
```

```python
import torch.nn.functional as F
class OneByOneNetowrk_v2(nn.Module):
    def __init__(self):
        super(OneByOneNetowrk_v2, self).__init__()
        self.flatten = nn.Flatten()
        self.linear1 = nn.Linear(28*28, 512)
        # self.relu1 = nn.ReLU()
        self.linear2 = nn.Linear(512, 512)
        # self.relu2 = nn.ReLU()
        self.linear3 = nn.Linear(512, 10)

    def forward(self, x):
        x = self.flatten(x)
        x = F.relu(self.linear1(x))
        x = F.relu(self.linear2(x))
        logits = self.linear3(x)
        return logits
```

```python
one_by_one_model_v2 = OneByOneNetowrk_v2().to(device)
print(one_by_one_model_v2)
```

#### (2) Sequential() 활용하기

```python
class SequentialNetwork(nn.Module):
    def __init__(self):
        super(SequentialNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.Dropout1d(),
            nn.ReLU(),
            nn.Linear(512, 10),
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
```

```python
sequential_network = SequentialNetwork().to(device)
print(sequential_network)
```

```python
class SequentialNetwork_v2(nn.Module):
    def __init__(self):
        super(SequentialNetwork_v2, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack1 = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU()
        )
        self.linear_relu_stack2 = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.Tanh(),
            nn.Linear(512, 512),
            nn.Tanh()
        )
        self.linear = nn.Linear(512*2, 10)

    def forward(self, x):
        x = self.flatten(x)
        out1 = self.linear_relu_stack1(x)
        out2 = self.linear_relu_stack2(x)
        out = torch.cat([out1, out2])
        logits = self.linear(out)
        return logits
```

```python
sequential_network_v2 = SequentialNetwork_v2().to(device)
print(sequential_network_v2)
```

#### (3) Model in Model

```python
class SubNetwork(nn.Module):
    def __init__(self):
        super(SubNetwork, self).__init__()
        self.hidden = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU()
        )

    def forward(self, x):
        out = self.hidden(x)
        return out
```

```python
class Model_in_Model(nn.Module):
    def __init__(self):
      super(Model_in_Model, self).__init__()
      self.flatten = nn.Flatten()
      self.hidden1 = SubNetwork()
      self.hidden2 = SubNetwork()
      self.fc = nn.Linear(512, 10)
    
    def forward(self, x):
      x = self.flatten(x)
      out1 = self.hidden1(x)
      out2 = self.hidden2(x)
      out = torch.cat([out1, out2])
      logit = self.fc(x)
      return logit
```

```python
model_in_model = Model_in_Model().to(device)
print(model_in_model)
```

### 4. 학습 후 모델 저장과 호출

#### (1) 데이터 준비

```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
```

```python
# 공개된 데이터 셋으로부터 학습 데이터 다운로드 
training_data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor(),
)

# 공개된 데이터 셋으로부터 테스트 데이터 다운로드
test_data = datasets.FashionMNIST(
    root="data",
    train=False,
    download=True,
    transform=ToTensor(),
)
```

```python
batch_size = 64

# Create data loaders.
train_dataloader = DataLoader(training_data, batch_size=batch_size)
test_dataloader = DataLoader(test_data, batch_size=batch_size)
```

#### (2) 모델

```python
class SequentialNetwork(nn.Module):
    def __init__(self):
        super(SequentialNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Dropout1d(),
            nn.Linear(512, 10),
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
```

```python
model = SequentialNetwork().to(device)
model
```

#### (3) Optimizer & Losses

```python
# Optimizer
optimizer = SGD(model.parameters(), lr=0.01, momentum=0.9)

# losses
losses = nn.CrossEntropyLoss()
```

#### (4) 학습 & 테스트 wrapper

```python
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    model.train()
    print("Training mode: {}".format(model.training))
    for batch, (X, y) in enumerate(dataloader):
        X, y = X.to(device), y.to(device)

        # Compute prediction error
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), batch * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
```

```python
def test(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()
    print("Training mode: {}".format(model.training))
    test_loss, correct = 0, 0
    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device), y.to(device)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

#### (5) 학습

```python
epochs = 5
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_dataloader, model, losses, optimizer)
    test(test_dataloader, model, losses)
print("Done!")
```
